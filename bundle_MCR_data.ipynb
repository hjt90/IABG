```python
import csv
import json
import os

dataset = 'food'
data_dir = os.path.join('data', dataset)
train_path = os.path.join(data_dir, 'user_bundle_item.csv')
test_path = os.path.join(data_dir, 'user_bundle_test.csv')
```

```python
# read csv
with open(train_path, 'r') as f:
    reader = csv.reader(f)
    data = list(reader)

print(data)
```

```python
user_bundle = {}
for row in data:
    user_id = row[0]
    bundle_id = int(row[1])
    bundle_item = row[2]
    bundle_item = bundle_item.split('|')
    bundle_item = [int(item) for item in bundle_item]
    if user_id not in user_bundle:
        user_bundle[user_id] = []
    user_bundle[user_id].append(bundle_item)
```

```python
print(user_bundle)
```

```python
# json
with open(os.path.join(data_dir,'bundle_MCR' , 'user_bundle_offine.json'), 'w') as f:
    json.dump(user_bundle, f)
```

```python
# read csv
with open(test_path, 'r') as f:
    reader = csv.reader(f)
    data = list(reader)

user_bundle = {}
for row in data:
    user_id = row[0]
    bundle_id = int(row[1])
    bundle_item = row[2]
    bundle_item = bundle_item.split('|')
    bundle_item = [int(item) for item in bundle_item]
    if user_id not in user_bundle:
        user_bundle[user_id] = []
    user_bundle[user_id].append(bundle_item)

# json
with open(os.path.join(data_dir,'bundle_MCR' , 'user_bundle_test.json'), 'w') as f:
    json.dump(user_bundle, f)

with open(os.path.join(data_dir,'bundle_MCR' , 'user_bundle_valid.json'), 'w') as f:
    json.dump(user_bundle, f)
```

```python
# read json
with open(os.path.join(data_dir,'processed' , 'new_item_ID_map.json'), 'r') as f:
    new_item_ID = json.load(f)

item_num = len(new_item_ID)
```

```python
item_id_lookup = {}

for item in range(item_num):
    item_id_lookup[str(item)] = str(item)

print(len(item_id_lookup))
print(item_id_lookup)

item_cate = {}

for item in range(item_num):
    item_cate[str(item)] = [0]

print(len(item_cate))
print(item_cate)

# json
with open(os.path.join(data_dir,'bundle_MCR' , 'item_id_lookup.json'), 'w') as f:
    json.dump(item_id_lookup, f)

with open(os.path.join(data_dir,'bundle_MCR' , 'item_cate.json'), 'w') as f:
    json.dump(item_cate, f)

with open(os.path.join(data_dir,'bundle_MCR' , 'item_attr.json'), 'w') as f:
    json.dump(item_cate, f)
```
