```python
dataset = 'clothing'
```

```python
# read csv
import pandas as pd
import os
bundle_intent = pd.read_csv(os.path.join('data', dataset, 'bundle_intent.csv'))
bundle_intent = bundle_intent['intent'].tolist()
print(bundle_intent)
```

```python
# read csv
user_bundle_item_history = pd.read_csv(os.path.join('data', dataset, 'user_bundle_item.csv'))
bundle_item = user_bundle_item_history.iloc[:, 1:3]
bundle_item = bundle_item.values.tolist()
print(bundle_item)
```

```python
# read json
import json
item_des = json.load(open(os.path.join('data', dataset, 'item_nlp_info_new_ID.json'), 'r'))
all_item = list(item_des.values())
print(item_des)
print(all_item)
```

```python
for i in range(len(bundle_item)):
    item_set = bundle_item[i][1]
    item_set = item_set.split('|')
    item_set = [item_des[item] for item in item_set]
    
    bundle_item[i] = [bundle_intent[i], item_set]

print(bundle_item)
```

```python
from transformers import AutoTokenizer, AutoModel
from utils import load_model_on_gpus

tokenizer = AutoTokenizer.from_pretrained("THUDM/chatglm-6b", trust_remote_code=True)
model = load_model_on_gpus("THUDM/chatglm-6b", num_gpus=2)
model = model.eval()
```

```python
import random
import numpy as np
K = 50
res = []
pre = []

for b_i in bundle_item:
    bundle_intent, item_des = b_i
    neg_item = list(set(all_item).difference(set(item_des)))
    neg_item = random.sample(neg_item, K - len(item_des))
    candidate = item_des
    candidate.extend(neg_item)
    shuffle_ix = np.random.permutation(len(candidate))
    candidate = candidate[shuffle_ix]
    ground_true = shuffle_ix[:len(item_des)]
    prompt = "I want {}, please choose 3-5 items as a bundle from ".format(bundle_intent)
    prompt += ', '.join(candidate)
    response, history = model.chat(tokenizer, prompt, history=[])
    print(response)
    res.append([pre, ground_true])
```
